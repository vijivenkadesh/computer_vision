Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

Loading an image

```python
image = cv2.imread('image28.jpg')
```

```python
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(2, 1, 1)
plt.imshow(image_grey)
plt.title(label='Grey Image')
plt.subplot(2, 1, 2)
plt.imshow(image_rgb)
plt.title(label='RGB Image')
plt.show()
```

Recognition

```python
stop_data = cv2.CascadeClassifier('stop_data.xml')
```

```python
dir(stop_data)
```

```python
found = stop_data.detectMultiScale(image_rgb, minSize=(20,20))
```

```python
for x, y, w, h in found:
    cv2.rectangle(image_rgb, (x, y), (x+h, y+w), (0, 255, 0), 5)
    plt.subplot(1, 1, 1)
    plt.imshow(image_rgb)
```
